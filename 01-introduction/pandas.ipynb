# DataFrame

```python
# create a dtafrane from two series
import pandas as pd
data = {
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df = pd.DataFrame(data)
print(df)
```

```python
# locate row using loc function
print(df.loc[0])
```

```python
# return 0 and 1
print(df.loc[[0,1]])
```

```python
# indexing with day 1 , day 2, day 3
import pandas as pd
data = {
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df = pd.DataFrame(data,index=["Day 1","Day 2","Day 3"])
print(df)
```

```python
df.loc["Day 2"]
print(df.loc["Day 1"])
```
